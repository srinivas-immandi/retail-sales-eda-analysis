# Finding Duplicated values

```python
df.duplicated().sum() 
```

# There is no duplicated rows in the dataset

# Finding Missing Values

```python
df.isnull().sum()
```

# OrderDate,Category,Quantity,ReviewText,Age These Columns Having The Missing Values

# Handling Null Values In The OrderDate Column

```python
a = df['OrderDate'].isnull().sum()
print('Null Values in OrderDate Column:',a)
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df['OrderDate'] = df['OrderDate'].bfill()
```

```python
a1 = df['OrderDate'].isnull().sum()
print('After Handling Null Values:',a1)
```

# Handling Null Values In The Category Column

```python
print('dtype of the Category:',df['Category'].dtype)
b = df['Category'].isnull().sum()
print('Null Values in Category Column:',b)
```

```python
b1 = df['Category'].mode()
print('Null values are Replaced by:',b1)
```

```python
df['Category'] = df['Category'].fillna(df['Category'].mode()[0])
```

```python
b2 = df['Category'].isnull().sum()
print('After Handling Null Values:',b2)
```

# Handling Null Values In The Quantity Column

```python
print('dtype of the Quantity:',df['Quantity'].dtype)

c = df['Quantity'].isnull().sum()

print('Null Values in Quantity Column:',c)
```

```python
c1 = df['Quantity'].median()
print('Null values are Replaced by:',c1)
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
```

```python
c3 = df['Quantity'].isnull().sum()
print('After Handling Null Values:',c3)
```

# Handling Null Values in the ReviewText Column

```python
print('dtype of the ReviewText:',df['ReviewText'].dtype)

d = df['ReviewText'].isnull().sum()

print('Null Values in ReviewText Column:',d)
```

```python
d1 = df['ReviewText'].mode()
print('Null values are Replaced by:',d1)
```

```python
df['ReviewText'] = df['ReviewText'].fillna(df['ReviewText'].mode()[0])
```

```python
d3 = df['ReviewText'].isnull().sum()
print('After Handling Null Values:',d3)
```

# Handling Null Values in the Age Column

```python
print('dtype of the Age:',df['Age'].dtype)

e = df['Age'].isnull().sum()

print('Null Values in Age Column:',e)
```

```python
e1 = df['Age'].median()
print('Null Values are Replaced by:',e1)
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
e2 = df['Age'].isnull().sum()
print('After Handling Null Values:',e2)
```

# Stacked Barplot on Gender vs ReviewText

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create crosstab
ct = pd.crosstab(df2['Gender'], df2['ReviewText'])

# Plot stacked bar chart
plt.figure(figsize=(12,6))
ct.plot(kind='bar', stacked=True, figsize=(12,6), edgecolor='black')

plt.title("Gender vs Review Text (Stacked Bar Plot)")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.legend(title="Review Text", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Line Chart on OrderDate,Sales

```python
import matplotlib.pyplot as plt

# Group by date to avoid messy lines
daily_sales = df2.groupby('OrderDate')['Sales'].sum()

plt.figure(figsize=(10,5))
plt.plot(daily_sales.index, daily_sales.values, color='blue')

plt.title("Daily Sales Trend(Line Chart)")
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
